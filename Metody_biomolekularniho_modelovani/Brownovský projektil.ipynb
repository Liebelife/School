```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
F = 0
G = -1
t_r = 0.1
range_t_r = 15
delta_t = 0.01
number_of_steps = int(range_t_r/delta_t)


X_0, Y_0 = 0, 0
V_0, W_0 = 5, 5

beta_choice = [0.3, 0.7, 2]
beta = beta_choice[0]
```

```python
V_dt = -t_r * V_0 * delta_t + F * delta_t + np.sqrt(beta**2 * delta_t) * np.random.normal(0, 1) + V_0
V_0, V_dt
```

```python
def generate_X(beta):
    V_0 = 5
    X_0 = 0
    X = [X_0]
    for i in range(number_of_steps+1):
        V_dt = -t_r * V_0 * delta_t + F * delta_t + np.sqrt(beta**2 * delta_t) * np.random.normal(0, 1) + V_0
        X.append(X[-1] + V_dt * delta_t)
        V_0 = V_dt
    return X

def generate_Y(beta):
    W_0 = 5
    Y_0 = 0
    Y = [Y_0]
    for i in range(number_of_steps+1):
        W_dt = -t_r * W_0 * delta_t + G * delta_t + np.sqrt(beta**2 * delta_t) * np.random.normal(0, 1) + W_0
        Y.append(Y[-1] + V_dt * delta_t)
        W_0 = W_dt
    return Y
```

```python
beta_choice
```

```python
for beta in beta_choice:
    plt.figure()
    for i in range(11):
        X = generate_X(beta)
        Y = generate_Y(beta)
        plt.plot(X,Y)
    plt.ylabel("Y")
    plt.xlabel("X")
    plt.title(f"Trajektorie brownovskho projektilu s beta = {beta}")
    plt.savefig(f"Trajektorie_beta={beta}.png")
```
