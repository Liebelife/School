```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_0 = 0
dt = 1
delta = 0.1
t_range = range(1,101)
```

```python
def print_trajectory(colors):
    plt.figure()
    for color in colors:
        X = np.zeros(100)
        for i, x in enumerate(X):
            X[i+1] = X[i] + np.sqrt(delta**2 * dt) * np.random.normal(0, 1)
            if i == len(X) - 2:
                break

        plt.scatter(t_range, X, s= 5, color = color)
        plt.plot(t_range, X, linewidth= 0.5, color= color)
        plt.title("trajektorie stic")
        plt.savefig("Trajektorie.png")
```

```python
print_trajectory(["red", "blue", "green"])
```

```python
def count_trajectory(number):
    trajectories = []
    mean = []
    sd = []
    for num in range(number):
        X = np.zeros(100)
        for i, x in enumerate(X):
            X[i+1] = X[i] + np.sqrt(delta**2 * dt) * np.random.normal(0, 1)
            if i == len(X) - 2:
                break
        trajectories.append(X)
    return trajectories
```
