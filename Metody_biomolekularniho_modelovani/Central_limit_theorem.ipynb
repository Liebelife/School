```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
import statistics
```

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
```

```python
for n in [10,20,50,100,200]:
    np.random.seed(40)
    p = 1/3
    fig, ax = plt.subplots(1, 1)
    k = np.arange(n+1)
    binom = stats.binom.pmf(k, n, p)
    print(np.mean(binom))
    plt.scatter(k, binom)
    
    x_axis = np.arange(0, k.max(), 0.01)
    
    x = np.random.binomial(k.max(), p, size=k.max())
    mean = statistics.mean(x)
    sd = statistics.stdev(x)

    plt.plot(x_axis, norm.pdf(x_axis, mean, sd), color="red")
    plt.title(f"n = {n}")
    plt.savefig("value{n}.png".format(n=n))
    plt.show()    
```
